REALIZANDO MUDANAS DE DADOS

```python
df['Total'] = df['Ataque'] + df['Atq Esp'] + df['Def Esp'] + df['Defensa'] + df['Velocidade']+ df['Vida']
df
```

```python
df.drop(columns=['Total']) # REMOVE COLUNA DA TABELA
```

FORMA ALTERNATIVA PARA CRIAR, POR EXEMPLO UMA COLUNA "TOTAL" DENTRO DA TABELA

```python
df['Total'] = df.iloc[:, 4:10].sum(axis = 1)
df
```

RENOMEANDO DADOS

```python
df.rename({'Defensa': 'Defesa'}, axis=1, inplace=True)
df
```

REORDENAR COLUNAS

```python
df = df[['Total', 'Vida', 'Defesa']]# Um dos modos para reogarnizar as colunas
```

```python
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] #TOTAL MUDA DE POSIO E PASSA AO TERCEIRO LOCAL DENTRO DA TABELA
df
```

```python
df.to_csv('modificado.csv') #SALVANDO O ARQUIVO
```

FILTRANDO DADOS

Tipo Fogo and "&" e Agua, Tipo Fogo or | Agua

```python
novo_df = df.loc[(df['Tipo 1'] == 'Fogo') & (df['Tipo 2'] == 'Agua')]
novo_df
```

```python
novo_df = df.loc[(df['Tipo 1'] == 'Fogo') | (df['Tipo 2'] == 'Agua')]
novo_df.reset_index(drop=True, inplace=True)
novo_df
```

FILTRANDO VERSOES MEGA

```python
novo_df = df.loc[~df['Nome'].str.contains('Mega ')]
novo_df
```

EXPRESSOES PARA PKEMONS DE AGUA E FOGO

```python
import re

novo_df = df.loc[df['Tipo 1'].str.contains('Fogo|Agua', flags=re.I, regex=True)]
novo_df
```

POKEMONS QUE COMEAM COM "PI"

```python
novo_df = df.loc[df['Nome'].str.contains('^pi', flags=re.I, regex=True)]
novo_df
```

MUDANAS CONDICIONAIS

```python
df.loc[df['Tipo 1'] == 'Fogueteiro', 'Tipo 1'] = 'Fogo'
df
```

```python
df.loc[df['Total'] > 500, 'Lendrio'] = True
df
```

```python
df.loc[df['Velocidade'] > 70, ['Lendrio', 'Geracao']] = ['Lendrio', 'Gerao']
df
```

PLOTAGEM DE DADOS

```python
#Histograma

import matplotlib.pyplot as plt

atributo = 'Defesa'
bins = range(0, 201, 20)
plt.xticks(bins)
df = pd.read_csv('modificado.csv')
plt.title(f'Distribuio de Pokemons por {atributo}')
plt.xlabel(f'Valor do atributo {atributo}')
plt.ylabel(f'Quantidade de Pokemons')
plt.hist(df[atributo], bins=bins)
plt.show()
```

GRFICO VERTICAL COM MATPLOTLIB

```python
atributos = 'Velocidade'
tipos = df.groupby('Tipo 1').mean().sort_values(atributo, ascending=False)
linha = tipos.index.tolist()
coluna = tipos[atributo].tolist()
figura = plt.figure(figsize=(14, 4)).add_axes([0,0,1,1])
figura.bar(linha, coluna)

plt.xlabel('Tipos de Pokemon')
plt.ylabel(f'Valor mdio do atributo {atributo}')
plt.title(f'ndices de Pokemons por {atributo}')

for i, v in enumerate(coluna):
  plt.annotate(f'{v:.2f}', (i-0.3, v + 1))

plt.show()
```
